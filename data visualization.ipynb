Data Plotting with pandas 

```python
import numpy as np
import pandas as pd
import matplotlib 
```

```python
diamonds = pd.read_csv('diamonds.csv')
diamonds.head(5)
```

```python
del diamonds['Unnamed: 0']
```

```python
diamonds.head(5)
```

```python
diamonds.hist(column='carat',color='orange')
```

```python
diamonds.hist(column='carat' , bins=50,color='orange',
               range=(0,3.5))
```

```python
diamonds['carat'].head()
```

```python
large = diamonds[diamonds['carat']>3.5] 
```

```python
large['carat'].describe()
```

```python
large['carat'].hist()
```

```python
diamonds.boxplot(column='carat')
```

```python
diamonds.boxplot(column='price', by='clarity')
```

```python
diamonds.boxplot(column="carat",        # Column to plot
                 by= "clarity",         # Column to split upon
                 figsize= (8,8));
```

```python
diamonds['carat'].plot(kind='density' , xlim=(0,5),
                       xlabel='carat')
```

```python
carat_table = pd.crosstab(index=diamonds['clarity'],columns='count')
carat_table
```

```python
carat_table.plot(kind='bar')
```

```python
carat_table = pd.crosstab(index=diamonds['clarity'],columns=diamonds['color'])

carat_table
```

```python
carat_table.plot(kind='bar')
```

```python
carat_table.plot(kind = 'bar',stacked='True')
```

```python
diamonds.plot(kind ='scatter',
              x= 'carat',y ='price')
```
